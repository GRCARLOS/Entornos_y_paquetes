# Comparación de modelos
## Matlab vs Python

Respuesta del modelo motor agragando un retardo de 1 segundo, generado a partir de una aproximación de PADE de 3 orden.

```python
import numpy as np
import matplotlib.pyplot as plt

ma=np.load('Data_m.npy')
py=np.load('Data_py.npy')

## Graficamos corriente
plt.figure()
plt.plot(ma[0][:], ma[1][:], label='Matlab')
plt.plot(py[0][:], py[1][:], label='Python')
plt.legend()
plt.title('Respuesta en corriente')
plt.xlabel('Tiempo (s)')
plt.ylabel('A')
plt.show()

# plt.plot(ma[0][0:20], ma[1][0:20], label='Matlab')
# plt.plot(py[0][0:20], py[1][0:20], label='Python')
```

```python
## Graficamos Posición
plt.figure()
plt.plot(ma[0][:], ma[2][:], label='Matlab')
plt.plot(py[0][:], py[2][:], label='Python')
plt.legend()
plt.title('Respuesta en Posición angular')
plt.xlabel('Tiempo (s)')
plt.ylabel('rad')
plt.show()
```

```python
## Graficamos Velocidad
plt.figure()
plt.plot(ma[0][:], ma[3][:], label='Matlab')
plt.plot(py[0][:], py[3][:], label='Python')
plt.legend()
plt.title('Respuesta en Velocidad angular')
plt.xlabel('Tiempo (s)')
plt.ylabel('rad/s')
plt.show()
```
